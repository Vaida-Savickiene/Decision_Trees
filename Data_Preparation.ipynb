#Binning Data

Binning can be really important in data exploration activity. It is typically used to transform continuous variables into discrete ones.

```
#get example datas
!wget -O df_titanic.csv https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv
```

```
df_titanic = pd.read_csv('df_titanic.csv')
from sklearn.utils import shuffle

# Shuffling the Dataset
df_titanic = shuffle(df_titanic, random_state = 42)
```

```
df_titanic.head()
```

```
bins = [0,4,17,65,99]
labels =['Toddler','Child','Adult','Elderly']

category = pd.cut(df_titanic['Age'], bins = bins, labels = labels)
```

```
#create a new column with bins
df_titanic.insert(2, 'Age Group', category)
```

```
df_titanic.head()
```

```
df_titanic['Age Group'].value_counts()
```

```
df_titanic.isnull().sum()
```
